## Drop unwanted columns

```python
#df2["Date"] = pd.to_datetime(df3["Last Update"].date())
#df2["date"]=pd.to_datetime(df3["Last Update"]).date()
#pd.to_datetime(df2['Last Update'], utc=False)

df2['Last_Update'] = pd.to_datetime(df2['Last Update']).dt.date
```

## Drop the old variables - Last Update and ObservationDate

```python
df3 = df2.drop(["Last Update","ObservationDate"], axis = 1 )
```

## Renaming and arranging the variables 

```python
df3.columns.values
```

```python
#df2["Last_Update"] = df2["Last Update"]
df3['Province_State'] = df3['Province/State']
df3['Country_Region'] = df3['Country/Region']

df4 = ['ObservedDate', 'Province_State', 'Country_Region',
       'Last_Update', 'Confirmed', 'Deaths', 'Recovered']
```

```python
df2 = df3[df4]
```

#### 3. Which Region contains maximum cases Confirmed on that single day?

```python
df3 = df2.groupby(['Country_Region'])['Confirmed'].max()
pd.set_option('display.max_rows', None)
```

```python
df3.sort_values(inplace = False)
```

#### Answer : 

Russia has maximum cases <strong>confirmed</strong> with the total <strong> 405843</strong> cases registed in a day
